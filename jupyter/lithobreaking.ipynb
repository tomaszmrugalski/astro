```python
# Imports
from astropy import units as u

from poliastro.bodies import Earth
from poliastro.twobody import Orbit

from perylune.orbit_tools import *
```

```python
# Care should be taken when defining elliptical orbits. The semi-major axis (a) is the maximum distance (apoapsis) from the perycenter of the system
# if eccentricity is greater than zero, the periapsis is is always lower. Here's an example of an invalid orbit:
# It has altitude of 300km (so apoapsis of 6678km) and eccentricity of 0.1. This gets the apoapsis of 6644km, i.e. 367km below Earth surface.
alt = 300*u.km
orb3 = Orbit.from_classical(Earth, Earth.R + alt, 0.1*u.one, 0.0*u.deg, 0*u.deg, 0*u.deg, 0*u.deg)
print_orb(orb3)
orb3.plot()
```
